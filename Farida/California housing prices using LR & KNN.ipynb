```python
import pandas as pd

housing_pd = pd.read_csv('/content/drive/MyDrive/Project/California Housing Price/dataset/housing.csv')
housing_pd
```

```python
housing_pd['ocean_proximity'].value_counts()
```

```python
#To take a random sample
housing_pd_shuffled = housing_pd.sample(n=len(housing_pd), random_state=1)
housing_pd_shuffled
```

```python
#convert dummy variables
pd.get_dummies(housing_pd_shuffled['ocean_proximity']).head()
```

```python
#Drop Categorical column "ocean_proximity"
housing_pd_shuffled.drop('ocean_proximity', axis=1).head()
```

```python
#Concatenate the housing_pd_shuffled with "ocean_proximity" without strings
housing_pd_final = pd.concat([housing_pd_shuffled.drop('ocean_proximity', axis=1),
                              pd.get_dummies(housing_pd_shuffled['ocean_proximity'])], axis=1)
housing_pd_final
```

```python
#List them with 'median_house_value' at the end

housing_pd_final = housing_pd_final[['longitude',	'latitude',
                                     'housing_median_age',	'total_rooms',
                                     'total_bedrooms',	'population',
                                     'households',	'median_income',
                                     '<1H OCEAN',
                                     'INLAND',	'ISLAND',
                                     'NEAR BAY',	'NEAR OCEAN', 'median_house_value']]
housing_pd_final
```

```python
#Drop the null values

housing_pd_final = housing_pd_final.dropna()
len(housing_pd_final)
```

```python
# Splitting the dataset into training, test and validation sets

train_pd, test_pd, val_pd = housing_pd_final[:18000], housing_pd_final[18000:19217], housing_pd_final[19215:]
len(train_pd), len(test_pd), len(val_pd)
```

```python
#Convert train_pd, test_pd and val_pd into numpy array

X_train, y_train = train_pd.to_numpy()[:, :-1], train_pd.to_numpy()[:, -1]
X_val, y_val = val_pd.to_numpy()[:, :-1], val_pd.to_numpy()[:, -1]
X_test, y_test = test_pd.to_numpy()[:, :-1], test_pd.to_numpy()[:, -1]

X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

```python
#Scaling and preprocessing the data

# Import StandardScaler to standarize every column with the same scale
from sklearn.preprocessing import StandardScaler
import numpy as np

#scaling from column "longitude" to "median_income"
scaler = StandardScaler().fit(X_train[:, :8])

#setting preprocessror function for scaling data
def preprocessor(X):
  # Create a copy of the input array X
  A = np.copy(X)
  #Apply scaling to the first 8 columns
  A[:, :8] = scaler.transform(A[:, :8])
  return A

X_train, X_val, X_test = preprocessor(X_train), preprocessor(X_val), preprocessor(X_test)
```

```python
X_train.shape, X_val.shape, X_test.shape
```

# **Using Linear Regression**

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression

#fit the trained X and y to the linear model (lm)
lm = LinearRegression().fit(X_train, y_train)
mse(lm.predict(X_train), y_train, squared=False), mse(lm.predict(X_val), y_val, squared=False)
```

**Validation error is much bigger than training error (Overfitting)**

# **Using KNN Regressor**

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=10).fit(X_train, y_train)
mse(knn.predict(X_train), y_train, squared=False), mse(knn.predict(X_val), y_val, squared=False)
```

Overfitting decreased

```python
mse_lm_train = mse(lm.predict(X_train), y_train)
mse_lm_test = mse(lm.predict(X_test), y_test)
RMSE_lm_test = mse(lm.predict(X_test), y_test, squared=False)

print("Linear Regression - MSE (Training):", mse_lm_train)
print("Linear Regression - MSE (Testing):", mse_lm_test)
print("Linear Regression - RMSE (Testing):", RMSE_lm_test)

mse_knn_train = mse(knn.predict(X_train), y_train)
mse_knn_test = mse(knn.predict(X_test), y_test)
RMSE_knn_test = mse(knn.predict(X_test), y_test, squared=False)

print("KNN Regressor - MSE (Training):", mse_knn_train)
print("KNN Regressor - MSE (Testing):", mse_knn_test)
print("KNN Regressor - RMSE (Testing):", RMSE_knn_test)
```
